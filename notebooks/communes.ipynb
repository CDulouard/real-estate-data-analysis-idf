```python
import folium

from data.domain.commune import Commune
```

```python
map_provider = "Cartodb Positron"
```

```python
communes = Commune.get_all(load_parents=True)
```

```python
communes_to_display = [commune for commune in communes if "departement" in commune.__dict__.keys() is not None and commune.departement.region.insee_code == "11"]
print(len(communes_to_display))
```

```python
france_center = [45, 2.35]

map_communes = folium.Map(location=france_center, zoom_start=6, tiles=map_provider)

for commune in communes_to_display:
    folium.GeoJson(
        commune.geojson_shape,
        name=commune.name,
        style_function=lambda x: {
            'fillColor': 'transparent',
            'fillOpacity': 0,
            'color': 'red',
            'weight': 2
        }
    ).add_to(map_communes)

folium.LayerControl().add_to(map_communes)
```
